# Function Definitions

```python
def read_fpaths(user, fpaths):
    upaths = filter(lambda x: x.name.split(".")[0].endswith(user), ALL_FPATHS)
    udict = dict(map(lambda x: (x.parent.name, pd.read_json(x)), upaths))
    return udict
```

# Data Reading

```python
# Defining and reading paths after sorting
RESPONSE = Path.cwd() / "EMA" / "response"
RESPONSE_DIRS = list(map(lambda x: x.name, RESPONSE.iterdir()))
ALL_FPATHS = list(filter(lambda x: "resp_time" in pd.read_json(x).columns, RESPONSE.rglob("*.json")))
ALL_USERS = set(map(lambda x: (x.name.split(".")[0].split("_")[-1]), ALL_FPATHS))
dir_wise_data = {dire: list(map(pd.read_json, filter(lambda x: x.parent.name==dire, ALL_FPATHS))) for dire in RESPONSE_DIRS}
data = {user: read_fpaths(user=user, fpaths=ALL_FPATHS) for user in ALL_USERS}
```

# Extracting useful EMA Data

```python
# Some directory wise statistics
print("Number of zero responses from users in each directories:")
for dire in RESPONSE_DIRS:
    num_res = (np.fromiter(map(len, dir_wise_data[dire]), dtype=int) < 1).sum() 
    print(f"{dire}: {num_res}")
```

```python
# sel_dires = ["Activity", "Class", "Sleep", "Social", "Exercise", "Stress"]
sel_dires = ["Stress", "Sleep"]
# sel_data =  {user: dict(filter(lambda x: x[0] in sel_dires, data[user].items())) for user in data.keys()}
sel_data={}
drop_cols = ["location", "null", "schedule"]
for user in data.keys():
    sel = list(filter(lambda x: x[0] in sel_dires, data[user].items()))
    sel = dict(map(lambda x:(x[0], x[1].drop(drop_cols, axis=1, errors="ignore")), sel))
    for k, v in sel.items():
        v["resp_time"] = pd.to_datetime(v["resp_time"]).dt.date
        sel[k] = v.set_index("resp_time")
    # cols = list(set(pd.concat(sel.values(), axis=1).columns))
    # empt = pd.DataFrame(columns=cols).set_index("resp_time")
    
    mer = reduce(lambda left, right: pd.merge(left=left, right=right, on="resp_time", suffixes=("_l", "_r"), how="outer"), sel.values())
    sel_data[user] = mer.dropna()
```
